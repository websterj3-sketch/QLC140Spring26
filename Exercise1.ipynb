This exercise will introduce python programming and plotting nutrient data using a python package called matplotlib.

```python
#First we create a cell that installs packages we'll be using
import pandas as pd
```

```python
#Start by importing the nutrient data by hand:
data = {
    "Depth_m": [0, 50, 100, 200, 500, 1000, 2000, 4000],
    "Nitrate_Atlantic_uM": [0.5, 1.2, 3.0, 8.5, 18.0, 24.0, 30.0, 32.0],
    "Nitrate_Pacific_uM": [0.6, 2.0, 6.0, 15.0, 28.0, 36.0, 42.0, 45.0],
}
df = pd.DataFrame(data)

df
```

```python
#Plot the data for the Atlantic in a Scatter plot. Make sure the 0 depth is at the top and 4000 m is at the bottom.
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.scatter(df['Nitrate_Atlantic_uM'], df['Depth_m'])
plt.xlabel('Nitrate_Atlantic_uM')
plt.ylabel('Depth (m)')
plt.title('Nitrate Concentration vs. Depth in Atlantic Ocean')
plt.gca().invert_yaxis() # Invert y-axis to have 0 at the top
plt.grid(True)
plt.show()
```

```python
#Do the same with the Pacific
#Creating a figure using matplotli
plt.figure(figsize=(8, 6))
#Now plot a scatter plot in this empty figure
plt.scatter(df['Nitrate_Pacific_uM'], df['Depth_m'])
#Create an x-axis label describing what is on the axis
plt.xlabel('Nitrate Pacific (uM)')
#Create a y-axis label describing what is on the axis
plt.ylabel('Depth (m)')
#Create a figure title
plt.title('Nitrate Profile in the Pacific Ocean')
# Invert y-axis to have 0 at the top
plt.gca().invert_yaxis()
plt.grid(True)
plt.show()
```

```python
#Now plot both in the same plot as two separate lines. Include a legend.
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 7))
plt.plot(df['Nitrate_Atlantic_uM'], df['Depth_m'], label='Atlantic Ocean')
plt.plot(df['Nitrate_Pacific_uM'], df['Depth_m'], label='Pacific Ocean')
plt.xlabel('Nitrate Concentration (uM)')
plt.ylabel('Depth (m)')
plt.title('Nitrate Concentration vs. Depth in Atlantic and Pacific Oceans')
plt.gca().invert_yaxis() # Invert y-axis to have 0 at the top
plt.grid(True)
plt.legend()
plt.show()
```
